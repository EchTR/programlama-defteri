```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# https://www.zenrows.com/blog/scraping-javascript-rendered-web-pages#installing-the-requirements
options = Options() 
options.headless = True
driver = webdriver.Chrome(options=options, executable_path=r"C:\Users\pc\Desktop\pypypy\web_scraping\chromedriver.exe")
driver.get("http://quotes.toscrape.com/js/")
time.sleep(10.5)
page_source = driver.page_source
driver.close()
```

```python
soup = BeautifulSoup(page_source, "lxml")
items = soup.find_all("div", class_="quote")
data = []
for item in items:
    tags = item.find_all("a", class_="tag")
    data.append([
        item.find("small").text,
        item.find("span", class_="text").text,
        [x.text for x in tags]
    ])
df = pd.DataFrame(data, columns=["Name", "Quote","Tags"])
df
```
