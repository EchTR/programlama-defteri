```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = "http://books.toscrape.com/catalogue/category/books/history_32/index.html"
```

```python
html_text = requests.get(url).text
soup = bs(html_text, "lxml")
```

```python
items = soup.find_all("article", class_="product_pod")
data = []
```

```python
#name = item.h3.text
#price = item.find("p", class_="price_color").text[1:]
#stock = item.find("p", class_="instock").text.strip()
#rating = item.find("p", class_="star-rating").attrs.get("class")[1]
```

```python
for item in items:
    rating = item.find("p", class_="star-rating").attrs.get("class")[1]
    if rating == "One": rating_new = 1
    elif rating == "Two": rating_new = 2
    elif rating == "Three": rating_new = 3
    elif rating == "Four": rating_new = 4
    elif rating == "Five": rating_new = 5
    data.append([ 
        item.h3.text, 
        item.find("p", class_="price_color").text[1:], 
        item.find("p", class_="instock").text.strip(), 
        rating_new])
```
